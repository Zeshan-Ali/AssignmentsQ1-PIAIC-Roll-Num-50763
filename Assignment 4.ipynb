
## Q 1
Your first assignment is to write code to get input from user and store in dictionary and then store that dictionary in list.
Here is the required input.
```python
[
    {"name" : "Hamza Ali",
     "age"  : 25,
     "work" : "A.I",
     "skills" : ["python", "tensorflow", "cloud"],
     "degree" : {"title" : "BS", "major" : "computer Science", "completionDate" : "Jun 2019"},
     "salary" : 259999
     },
     {"name" : "Farooq Aziz",
     "age"  : 27,
     "work" : "Cloud Native",
     "skills" : ["ubuntu", "docker", "kubernetes"],
     "degree" : {"title" : "MS", "major" : "Software engineer", "completionDate" : "May 2018"},
     "salary" : 200000
     },
     {"name" : "Abu Bakr",
     "age"  : 30,
     "work" : "Blockchain",
     "skills" : ["python", "Cryptography", "Ethereum"],
     "degree" : {"title" : "BS", "major" : "computer Science", "completionDate" : "May 2019"},
     "salary" : 300000
     }
]
```

````python
main_list = []
# empty list to store the record entered by user

iteration = int(input("how many records you want to enter" ))
for i in range(iteration):
    myDic = {}
   # itera = int(input("Enter how many values you want to enter: "))
    for i in range(6):
        if i==0:
            name = input("enter your Name:")
            myDic["Name"] = name
        if i==1:
            age = input("enter your age: ")
            myDic["age"] = age
        if i==2:
            work = input("enter your work: ")
            myDic["work"] = work
        if i==3:
            skills = []  # empty list to store the skill entered by user
            skill = input("Enter your skills seprated by commas:")
            #skill = skill.split(",")
            skills.append(skill)
            myDic["Skills"] = skills
        if i==4:
            degree ={}
            for d in range(3):
                if d ==0:
                    title = input("Enter your degree title: ")
                    degree["Title"] = title
                if d ==1:
                    major = input("Enter your degree Major: ")
                    degree["Major"] = major
                if d ==2:
                    cmplDate = input("Enter your degree Completion data: ")
                    degree["CompletionDate"] = cmplDate
            myDic["degree"] = degree
        if i==5:
            salary = input("enter your salary: ")
            myDic["salary"] = salary
    main_list.append(myDic)
````
